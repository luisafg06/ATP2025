```python
#exercicio 1
#a
def contaespaços(string):
    if string == '':
        return 0
    if string[0] == '':
        return 1 + contaespaços(string[1:])
    else:
        return contaespaços(string[1:])
    
#b
def ordenada(lista):
    if len(lista)  <= 1 :
        return True
    if lista[0]>lista[1]:
        return False
    return ordenada(lista[1:])

#c
def existeElem(lista, n):
    if lista == []:
        return False
    if lista[0] == n:
        return True
    return existeElem(lista[1:],n)

#exxercicio 2
#a
def maiorDivisor(N):
    dic={}
    for num in range(1, N+1):
        maior=1
        for d in range(1,10):
            if num%d == 0:
                maior=d
        dic[num] = maior
    return dic 

#b
def mmc(a, b):
    mmc = a
    while mmc % b != 0:
        mmc += a
    return mmc

#c
def f2c(fahrenheit):
    novo = {}
    for k in fahrenheit:
        F = fahrenheit[k]
        C = (F - 32) / 1.8
        novo[k] = (F, C)
    return novo

#d
def reduxInt(n):
    while n >= 10:
        soma = 0
        for d in str(n):
            soma += int(d)
        n=soma
    return n

#exercicio 3

hospital = [
 (
 "Unidade de Cardiologia",
 [ (1, "Dr. João"), (2, "Dra. Maria")],
 [(101, False), (102, True), (103, False), (104, False)]
 ),
 (
 "Unidade de Pediatria",
 [ (3, "Dr. Pedro"), (4, "Dra. Clara") ],
 [ (201, True), (202, True), (203, True)]
 )
]

#a
def listar_unidades(hospital):
    nomes=[]
    for unidade in hospital:
        nomes.append(unidade[0])
    return nomes
print(listar_unidades(hospital))

#b
def camas_disponiveis(hospital):
    res = 0
    for unidade in hospital:
        camas = unidade[2]
        for cama in camas:
            if cama[1] == False:
                res= res + 1
    return res
print(camas_disponiveis(hospital))

#c
def listar_medicos_unidade(hospital, nome_unidade):
    for unidade in hospital:
        if unidade[0] == nome_unidade:
            medicos = []
            for medico in unidade[1]:
                medicos.append(medico)
            return medicos
    return []
print(listar_medicos_unidade(hospital, "Unidade de Cardiologia"))

#d
def existe_medico(hospital, nome_unidade, nome_medico):
    for unidade in hospital:
        if unidade[0] == nome_unidade:
            for medico in unidade[1]:
                if medico[1] == nome_medico:
                    return True
            return False
    return False

print(existe_medico(hospital, "Unidade de Cardiologia", "Dra. Maria"))

#e
def adicionar_medico(hospital, nome_unidade, nome_medico):
    for unidade in hospital:
        if unidade[0] == nome_unidade:
            for medico in unidade[1]:
                if medico[1] == nome_medico:
                    return hospital   
                else:
                    novo_id = len(unidade[1]) + 1
                    unidade[1].append((novo_id, nome_medico))
                return hospital
    return hospital
print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra. Paula"))

#f
def ocupar_cama(hospital, nome_unidade, id_cama):
    for i in range(len(hospital)):
        unidade = hospital[i]
        if unidade[0] == nome_unidade:
            camas = unidade[2]
            novas_camas = []
            for cama in camas:
                if cama[0] == id_cama and cama[1] == False:
                    novas_camas.append((cama[0], True))   
                else:
                    novas_camas.append(cama)
            hospital[i] = (unidade[0], unidade[1], novas_camas)
            return hospital
    return hospital
print(ocupar_cama(hospital, "Unidade de Cardiologia", 101))

#g
def guardar_hospital(hospital, nome_ficheiro):
    f = open(nome_ficheiro, "w")
    f.write(str(hospital))
    f.close()

#h
def carregar_hospital(nome_ficheiro):
    f = open(nome_ficheiro, "r")
    hospital = eval(f.read())
    f.close()
    return hospital

#i
def camas_ocupadas(hospital):
    total = 0
    ocupadas = 0

    for unidade in hospital:
        for cama in unidade[2]:
            total = total + 1
            if cama[1] == True:
                ocupadas = ocupadas + 1

    if total == 0:
        return 0
    return (ocupadas / total) * 100

#j
def unidades_ocupacao_critica(hospital):
    res = []

    for unidade in hospital:
        total = 0
        ocupadas = 0

        for cama in unidade[2]:
            total = total + 1
            if cama[1] == True:
                ocupadas = ocupadas + 1

        if total > 0:
            perc = (ocupadas / total) * 100
            if perc >= 80:
                res.append((unidade[0], perc))

    return res
```
